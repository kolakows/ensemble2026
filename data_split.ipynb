```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load full training data
df = pd.read_parquet('chebi_dataset_train.parquet')
print(f"Full dataset: {len(df)} molecules")

# Split 80/20
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)
print(f"Train: {len(train_df)}, Val: {len(val_df)}")

# Save splits
train_df.to_parquet('data/train.parquet', index=False)
val_df.to_parquet('data/val.parquet', index=False)
print("Saved to data/train.parquet and data/val.parquet")
```
